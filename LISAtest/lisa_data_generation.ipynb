```python
import matplotlib.pyplot as plt
import numpy as np
import math
import time 
import matplotlib as mpl
import os 
import scipy.signal as sig
from my_signal import power_law
from utils import cent_series as cent
```

```python
#constants
arm_length = 2.5e9  
speed_of_light = 299792458  
f_ref = speed_of_light / (4 * arm_length)  
f1 = 4e-4
Np = 4e-41
Na = 1.44e-48
h0=2.7e-18
pi=np.pi
```

```python
## Definition function
def cos2(x):
    return np.cos(x)**2

def SI(Na,f):
    return 4*Na*(1+(f1/f))

def SII(Np):
    return Np

def SA(Na,f):
    return SI(Na,f)/4*(2*pi*f)**-4

def Ss(Np):
    return SII(Np)
    
def N1(Np,Na,f):
    W = abs(1-np.exp(-2j*f/f_ref))
    return (4*Ss(Np)+8*(1+cos2(f/f_ref))*SA(Na,f))*W*W

def N2(Np,Na,f):
    W = abs(1-np.exp(-2j*f/f_ref))
    return -(2*Ss(Np)+8*SA(Na,f))*np.cos(f/f_ref)*W*W

def NA(Np,Na,f):
    return 2/3*(N1(Np,Na,f) - N2(Np,Na,f))

def NT(Np,Na,f):
    return 1/3*abs(N1(Np,Na,f) + 2*N2(Np,Na,f))

def perio(y_c):
    return(sig.csd(y_c, y_c, fs = 1, nperseg=int(len(y_c)))[1])

#LISA response to GW
def response(f):
    W_sq = (abs(1-np.exp(-2j*f/f_ref)))**2
    R_A=9*W_sq*(1/(1+f/(4*f_ref/3)))/20
    return R_A

def psd_sgwb(omega,response,f):
    return(3*(h0**2)*omega*response)/(4*(pi**2)*f**3)

```

```python
#Adding a simple power law to the noise (blocked)
#True noise PSD
# Time domain parameters
obs_time = 10* 24 * 60 * 60  # Total time in seconds (ten days)
dt = 1  # Time step in seconds
t = np.arange(0, obs_time, dt)
N_t = len(t)
freq = np.fft.rfftfreq(N_t, dt)[1:]
T=NT(Np,Na,freq)
# Generate random noise in the frequency domain
N_f = len(freq)
noise_f_real = np.random.normal(0, np.sqrt(N_t * T / (4 * dt)), N_f)
noise_f_imag = np.random.normal(0, np.sqrt(N_t * T / (4 * dt)), N_f)
noise_f = noise_f_real + 1j * noise_f_imag

# Transform noise to the time domain
noise_t = np.fft.irfft(noise_f)

#power law variation
b=np.log(1e-46)
g=-2/3
sigpsd =np.exp(power_law(b,g,freq))
sig_f_real = np.random.normal(0, np.sqrt(N_t * sigpsd / (np.pi * dt)), N_f)
sig_f_imag = np.random.normal(0, np.sqrt(N_t * sigpsd / (np.pi * dt)), N_f)
sig_f = sig_f_real + 1j * sig_f_imag
sigtimeser = np.fft.irfft(sig_f)

timeser=noise_t+sigtimeser
sdser=np.std(timeser)
y_c=cent(timeser)
```

```python
#blocking
n_bl=10
bl_size=len(y_c)//n_bl
y_c = np.array(y_c[:bl_size * n_bl])
matrix = y_c.reshape(n_bl, -1)

T_per=np.array([perio(row) for row in matrix])
f=np.linspace(0,0.5,len(T_per[0]))
cond= (f >= 1e-4) & (f <= 1e-1)
f = f[cond]
T_per = T_per[:, cond]
T=NT(Np,Na,f)
sigpsd=np.exp(power_law(b,g,f))
tpsd=T+sigpsd
```

```python
plt.plot(f, T_per[0]*(sdser)**2, linestyle='-', color='black', alpha=0.4,label='T channel')
plt.plot(f, tpsd, linestyle='--', color='black', label='True PSD ')
plt.xscale('log')
plt.yscale('log')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize='small')
```

```python
#TDI with signal
obs_time = 2*24 * 60 * 60  # Total time in seconds (One day)
dt = 1  # Time step in seconds
t = np.arange(0, obs_time, dt)
N_t = len(t)
freq = np.fft.rfftfreq(N_t, dt)[1:]

#T channel
T=NT(Np,Na,freq)

#A,E channel noise
A=NA(Np,Na,freq)
E=A

#GW signal 
b=np.log(1e-13)
g=-2/3
sigpsd =np.exp(power_law(b,g,freq))
#response:
res=response(freq)
#with LISA response:
signal=psd_sgwb(omega=sigpsd,response=res,f=freq)
cond=cond= (freq >= 1e-4) & (freq <= 1e-1)

f_1=freq[cond]
signal_1=signal[cond]
A_1=A[cond]
#SNR (Smith and Cadwell formula after I approximated the integral using mid-point rule)
del_f=(max(f_1)-min(f_1))/len(f_1)
integ=del_f*np.sum(signal_1**2/A_1**2)
snr=np.sqrt(obs_time*integ)
print(snr)
```

```python
#blocking
n_bl=2
bl_size=len(y_c_T)//n_bl

#A channel
y_c_A = np.array(y_c_A[:bl_size * n_bl])
matrix = y_c_A.reshape(n_bl, -1)
A_per=np.array([perio(row) for row in matrix])

#E channel
y_c_E = np.array(y_c_E[:bl_size * n_bl])
matrix = y_c_E.reshape(n_bl, -1)
E_per=np.array([perio(row) for row in matrix])

#T channel
y_c_T = np.array(y_c_T[:bl_size * n_bl])
matrix = y_c_T.reshape(n_bl, -1)
T_per=np.array([perio(row) for row in matrix])


f=np.linspace(0,0.5,len(A_per[0]))
cond= (f >= 1e-4) & (f <= 4e-2)
f = f[cond]
A_per = A_per[:, cond]
E_per = E_per[:, cond]
T_per = T_per[:, cond]
sigpsd =np.exp(power_law(b,g,f))
res=response(f)
signal=psd_sgwb(omega=sigpsd,response=res,f=f)
A=NA(Np,Na,f)
E=A
T=NT(Np,Na,f)
```

```python
def inverse_transform(A, E, T):
    X = 0.5*(2*T/np.sqrt(3) - np.sqrt(2)*A + np.sqrt(6)*E/3)
    Y = (1/3)*(np.sqrt(3)*T - np.sqrt(6)*E)
    return X, Y
```
